```python
import torch 
from torch import nn 
import numpy as np 
import torchvision
from torch.utils.data import DataLoader, Dataset
import matplotlib.pyplot as plt 
import matplotlib
import torch.nn.functional as F
import tqdm
from IPython.display import clear_output
matplotlib.style.use("default")
```

```python
mnist_data = torchvision.datasets.MNIST("MNIST_DATA/",
                                  train= True,
                                  transform= torchvision.transforms.ToTensor(),
                                  download = True
                                 )

mnist_data = mnist_data.data/255.0
```

```python
def plot_sample(X, img_shape, sample_size = 10, n_rows = 2, n_columns = 5):
    if sample_size != n_rows*n_columns:
        raise Exception("the argument sample_size should equal n_rows * n_columns")
    for i in range(sample_size):
        if i%(sample_size//2) == 0:
            figure = plt.figure(figsize=(15,15))
        ax = plt.subplot(n_rows, n_columns, i+1)
        idx = torch.randint(low = 0, high =len(X), size = (1,))
        plt.imshow(X[idx].view(*img_shape), cmap='gray')
        
def plot_all(X, img_shape, n_rows = 2, n_columns = 5):
    sample_size = len(X)
    if sample_size != n_rows*n_columns:
        raise Exception("the argument sample_size should equal n_rows * n_columns")
    for i in range(sample_size):
        if i%(sample_size//2) == 0:
            figure = plt.figure(figsize=(15,15))
        ax = plt.subplot(n_rows, n_columns, i+1)
        plt.imshow(X[i].view(*img_shape), cmap='gray')
        
plot_sample(mnist_data, (28,28))
```

```python
batch_size= 512
data_loader = DataLoader(mnist_data, batch_size= batch_size)
```

```python
class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        self.fc1 = nn.Linear(in_features= 28*28, out_features= 512)
        self.fc2 = nn.Linear(in_features= 512, out_features= 256)
        self.fc3 = nn.Linear(in_features= 256, out_features= 128)
        self.fc4 = nn.Linear(in_features= 128, out_features= 1)
        self.dropout = nn.Dropout(0.35)
        
    def forward(self, x):
        out = self.dropout(F.leaky_relu(self.fc1(x), negative_slope=0.2))
        out = self.dropout(out)
        out = self.dropout(F.leaky_relu(self.fc2(out), negative_slope=0.2))
        out = self.dropout(out)
        out = self.dropout(F.leaky_relu(self.fc3(out), negative_slope=0.2))
        out = self.dropout(out)
        out = self.fc4(out)
        return out
```

```python
class Generator(nn.Module):
    def __init__(self):
        super(Generator, self).__init__()
        self.fc1 = nn.Linear(in_features= 128, out_features= 64)
        self.fc2 = nn.Linear(in_features= 64, out_features= 128)
        self.fc3 = nn.Linear(in_features= 128, out_features= 256)
        self.fc4 = nn.Linear(in_features= 256, out_features= 512)
        self.fc5 = nn.Linear(in_features= 512, out_features= 28*28)
        self.dropout = nn.Dropout2d(0.35)
    def forward(self, x): 
        out = F.relu(self.fc1(x))
        out = self.dropout(out)
        out = F.relu(self.fc2(out))
        out = self.dropout(out)
        out = F.relu(self.fc3(out))
        out = self.dropout(out)
        out = F.relu(self.fc4(out))
        out = self.dropout(out)
        out = torch.tanh(self.fc5(out))
        return out
```

```python
gen = Generator()
disc = Discriminator()

gen = gen.cuda()
disc = disc.cuda()
```

```python
disc_opt = torch.optim.Adam(disc.parameters())
gen_opt = torch.optim.Adam(gen.parameters())
```

```python
def real_loss(disc_out, smoothing = False):
    targets = torch.ones(disc_out.size(0)) * 0.9 if smoothing else torch.ones(disc_out.size(0)) 
    targets = targets.cuda()
    criterion = nn.BCEWithLogitsLoss()
    loss = criterion(torch.squeeze(disc_out), targets)
    return loss

def fake_loss(disc_out):
    targets = torch.zeros(disc_out.size(0))
    targets = targets.cuda()
    criterion = nn.BCEWithLogitsLoss()
    loss = criterion(torch.squeeze(disc_out), targets)
    return loss
```

```python
epochs = 1000
AdvLoss = [] 
DiscLoss = [] 
samples = [] 
```

```python
for epoch in tqdm.tqdm(range(epochs), total = epochs):
    for b, batch in enumerate(data_loader):
        batch = (batch *2 -1).view(-1, 28*28)
        batch = batch.cuda()
        disc_opt.zero_grad()
        disc_real_out = disc(batch)
        real_disc_loss = real_loss(disc_real_out, smoothing = True)
        Z = torch.from_numpy(np.random.uniform(low=-1, high=1, size=(batch_size, 128))).float().cuda()
        fake_data = gen(Z)
        disc_fake_out = disc(fake_data)
        fake_disc_loss = fake_loss(disc_fake_out)
        total_loss = fake_disc_loss + real_disc_loss
        total_loss.backward()
        disc_opt.step()
        
        gen_opt.zero_grad()
        Z = torch.from_numpy(np.random.uniform(low=-1, high=1, size=(batch_size, 128))).float().cuda()
        fake_data = gen(Z)
        disc_fake_out = disc(fake_data)
        adv_loss = real_loss(disc_fake_out)
        adv_loss.backward()
        gen_opt.step()
    clear_output()
    print(f"Epoch: {epoch+1}/{epochs}, adv_loss: {adv_loss}, disc_loss: {total_loss}")
    gen.eval()
    Z = torch.from_numpy(np.random.uniform(low=-1, high=1, size=(10, 128))).float().cuda()
    generated_sample = gen(Z)
    samples.append(generated_sample)
    gen.train()
    AdvLoss.append(adv_loss)
    DiscLoss.append(total_loss)
```

```python
torch.save(gen, "Generator.pt")
torch.save(gen, "Discriminator.pt")
```

```python
figure = plt.figure(figsize = (10,5))
plt.plot(AdvLoss, label = "Adv_Loss")
plt.plot(DiscLoss, label = "Disc_Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
```
